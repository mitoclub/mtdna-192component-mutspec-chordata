```python
from lxml import etree
import pandas as pd
```

```python
# Define needed taxes
taxa = {'order', 'class', 'family', 'genus', 'species'}
# Parse our file
root = etree.parse('../data/taxonomy_result.xml')
all_sps_taxa = []
for species_record in root.findall('Taxon'):
    sps_taxa = {}
    for taxes in species_record.find('LineageEx').findall('Taxon'):
        rank = taxes.find('Rank').text
        sc_name = taxes.find('ScientificName').text
        if rank in taxa:
            sps_taxa[rank] = sc_name
            
    all_sps_taxa.append(sps_taxa)
```

```python
final_df = pd.DataFrame(all_sps_taxa)
```

### Drop NA and duplicates from df

```python
final_df = final_df.dropna()
```
